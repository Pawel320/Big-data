### <a name='0'></a> Import bibliotek

```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

### <a name='1'></a> Wygenerowanie danych

```python
data = {
    'size':['XL','L','M','L','M'],
    'color':['red','green','blue','green','red'],
    'gender':['female','male','male','female','female'],
    'price':[199.0, 89.0, 99.0, 129.0, 79.0],
    'weight':[500, 450, 300, 380, 410],
    'bought':['yes','no','yes','no','yes']
}
df_raw = pd.DataFrame(data=data)
df_raw
```

### <a name='2'></a> Utworzenie kopii danych

```python
df = df_raw.copy()
df.info()
```

```python
for col in ['size','color','gender','bought']:
  df [col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')
df.info()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

### <a name='3'></a> LabelEncoder

```python
le.classes_
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

### <a name='6'></a> Standaryzacja - StandardScaler

std() - pandas nieobciążony

std() - numpy obciążony

```python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie std: {df['price'].std():.4f}")
```

```python
(df['price']- df['price'].mean())/df['price'].std()
```

```python
def standarize(x):
  return (x-x.mean())/x.std()
```

```python
print(df['price'])
```

```python
print(standarize(df['price']))
```

```python
df['price'] = standarize(df['price'])
df['price']
```

```python
#to samo tylko Z biblioteki sklearn
from sklearn.preprocessing import scale
scale(df['price'])
# jedyna różnica jest taka że jeden  używa std obciązonego a drugi nieobciazonego
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```

```python
scaler = StandardScaler()
df[['price','weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
